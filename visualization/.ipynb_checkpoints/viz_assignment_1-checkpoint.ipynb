* **value** : data value for element (tenths of degrees C)

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import pandas as pd


df_raw = pd.read_csv('./fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')

df_raw.head()
```

```python
import datetime


df = df_raw.copy()
df.rename({'Date': 'Date_Str', 'Data_Value': 'Temperature (C*10)'}, axis=1, inplace=True)
df['Date'] = pd.to_datetime(df.loc[:, 'Date_Str'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Month-Day'] = df['Date'].dt.strftime('%m-%d')
df['Temperature (C)'] = df['Temperature (C*10)'].copy() / 10
df = df[~((df['Month'] == 2) & (df['Day'] == 29))].copy()  # Getting Rid of leap days

df
```

```python
df.drop(['Date_Str', 'Temperature (C*10)'], axis=1, inplace=True)

df
```

```python
df_tmax = df.set_index('Element').sort_index().loc['TMAX'].reset_index().drop('Element', axis=1).copy()
df_tmin = df.set_index('Element').sort_index().loc['TMIN'].reset_index().drop('Element', axis=1).copy()

df_tmax
```

```python
df_tmin
```

```python
df_tmax_2015 = df_tmax[df_tmax['Year'] == 2015].copy()
df_tmax_10yr = df_tmax[~(df_tmax['Year'] == 2015)].copy()
df_tmax_2015.set_index('Month-Day', inplace=True)
df_tmax_2015.sort_index(inplace=True)
df_tmax_2015
```

```python
df_tmax_10yr
```

```python
df_tmin_2015 = df_tmin[df_tmin['Year'] == 2015].copy()
df_tmin_10yr = df_tmin[~(df_tmin['Year'] == 2015)].copy()
df_tmin_2015.set_index('Month-Day', inplace=True)
df_tmin_2015.sort_index(inplace=True)

df_tmin_2015
```

```python
df_tmin_10yr
```

```python
df_tmax_10yr_agg = df_tmax_10yr.loc[df_tmax_10yr.groupby(['Month', 'Day'], as_index=False).idxmax()['Temperature (C)']].copy()
df_tmax_10yr_agg.set_index('Month-Day', inplace=True)

df_tmax_10yr_agg
```

```python
df_tmin_10yr_agg = df_tmin_10yr.loc[df_tmin_10yr.groupby(['Month', 'Day'], as_index=False).idxmin()['Temperature (C)']].copy()
df_tmin_10yr_agg.set_index('Month-Day', inplace=True)

df_tmin_10yr_agg
```

```python
df_tmax_2015_agg = df_tmax_2015.groupby(level=0).max()
df_tmax_2015_filter = df_tmax_2015_agg[df_tmax_2015_agg['Temperature (C)'] > df_tmax_10yr_agg['Temperature (C)']]

df_tmax_2015_filter
```

```python
df_tmin_2015_agg = df_tmin_2015.groupby(level=0).min()
df_tmin_2015_filter = df_tmin_2015_agg[df_tmin_2015_agg['Temperature (C)'] < df_tmin_10yr_agg['Temperature (C)']]

df_tmin_2015_filter
```

```python
# Converting all indices back to datetime format

def conv_df(df):
    df_new = df.reset_index()
    df_new['Month-Day-dt'] = pd.to_datetime(df_new['Month-Day'].copy(), format='%m-%d')
    df_new.set_index('Month-Day-dt', inplace=True)
    df_new.drop('Month-Day', axis=1, inplace=True)
    return df_new

df_lst = [df_tmax_10yr_agg, df_tmin_10yr_agg, df_tmax_2015_filter, df_tmin_2015_filter]
df_out = []
for df in df_lst:
    df_out.append(conv_df(df))
    
df_tmax_10yr_final, df_tmin_10yr_final, df_tmax_2015_final, df_tmin_2015_final = df_out[0], df_out[1], df_out[2], df_out[3]

df_tmax_10yr_final
```

```python
import matplotlib.dates as mdates
from matplotlib.ticker import EngFormatter, StrMethodFormatter

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

tmax_10yr = df_tmax_10yr_final['Temperature (C)'].copy()
tmin_10yr = df_tmin_10yr_final['Temperature (C)'].copy()
tmax_2015 = df_tmax_2015_final['Temperature (C)'].copy()
tmin_2015 = df_tmin_2015_final['Temperature (C)'].copy()
tmax = plt.plot(tmax_10yr.index,
                tmax_10yr.values,
                color='red',
                alpha=.40,
                linewidth=2,
                label='Max Temp 2005-2014',
                )
tmin = plt.plot(tmin_10yr.index,
                tmin_10yr.values,
                color='blue',
                alpha=.40,
                linewidth=2,
                label='Min Temp 2005-2014'
                )

plt.scatter(tmax_2015.index, 
            tmax_2015.values,
            color='none',
            edgecolors='red',
            marker='^',
            s=150,
            linewidths=1.5,
            label='Max Temp 2015'
           )
plt.scatter(tmin_2015.index, 
            tmin_2015.values,
            color='none',
            edgecolors='blue',
            marker='v',
            s=150,
            linewidths=1.5,
            label='Min Temp 2015')
plt.legend(loc='lower right', frameon=False, prop=dict(size=12))

l1, l2 = ax.get_lines()
x, y = l1.get_data()
y1 = l2.get_data()[1]
ax.fill_between(x, y, y1, color='purple', alpha=0.1)

ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.set_xlim(tmax_10yr.index.min(), tmax_10yr.index.max())

ax.yaxis.set_major_formatter(EngFormatter(unit=u"°C"))
ax.set_ylim(-45, 45)

for dir in ['top', 'bottom', 'left', 'right']:
    ax.spines[dir].set_visible(False)
    
ax.tick_params(top=False, bottom=False, left=False, right=False, labelcolor ='black', labelsize=14)

plt.title('Temperature Trends at Ann Arbor, MI and Surrounding Areas', fontsize=16)
```
