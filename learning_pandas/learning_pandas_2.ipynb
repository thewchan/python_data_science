### Series Plotting

The Series object in pandas contains build-in plotting methods utilzing the matplotlib package. Specifically, the `.plot()` method plots the index against the values. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

songs_69 = pd.Series([18, 22, 7, 5],
                    index=['John', 'Paul', 'Geroge', 'Ringo'],
                    name='Counts 69')

songs_66 = pd.Series([5.0, 7.0, 18.0, 22.0, np.nan],
                    index=['Ringo', 'George', 'John', 'Paul', 'Eric'],
                    name='Counts 66')

songs_69
```

```python
songs_66
```

```python
fig = plt.figure()
songs_69.plot()
songs_66.plot()
plt.legend() # This funciton will insert a legend in the current plot
```

By default, `.plot()` creates a line chart, but using the `kind=` parameter we can create bar charges as well. By default, the bars will not stack, but cover one another. We can adjust the color of the lines/bars with the `color=` parameter and their transparency by the `alpha=` parameter. The transparency can help us see both bars even though they cover each other:

```python
fig = plt.figure()
songs_69.plot(kind='bar')
songs_66.plot(kind='bar', color='black', alpha=.5)
plt.legend()
```

Generating a histogram is also simiple in pandas by using the `.hist()` method:

```python
data = pd.Series(np.random.randn(500),
                name='500 random')
fig = plt.figure()
ax = fig.add_subplot(111)
data.hist()
```

Compare this to a matplotlib histogram:

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(data)
```

We can plot a kernel density estimation plot (KDE), which is like a histogram but uses a curve instead of bins and bars (this requires the scipy.stats package):

```python
fig = plt.figure()
data.plot(kind='kde')
```

Because pandas plotting is essentially built on top of matplotlib, much of the functionality from matplotlib would also work with pandas. For example, we can add two more plots to the figure like so:

```python
fig = plt.figure()
ax = fig.add_subplot(211)
data.plot(kind='kde',
         color='black',
         alpha=.6,
         ax=ax)
data.hist(color='green',
         alpha=.6,
         ax=ax,
         density=True)
ax.set_title('KDE, Histogram, and CDF')
ax = fig.add_subplot(212)
data.hist(ax=ax,
         density=True,
         cumulative=True)
```

The pandas Series (and DataFrame) object `.plot()` can plot many kinds of chart by passing diffrerent `kind=` parameter, or as attributes of the `.plot` attribute:

* `plot.area`: Creates an area plot for numeric columns
* `plot.bar`: Creats a bar plot for numberic columns
* `plot.barh`: Creates a horizontal bar plot for numeric columns
* `plot.box`: Creates a box plot for numeric columns
* `plot.density`: Creats a kernel density estimation plot for numeric columns (also `plot.kde`)
* `plot.hist`: Creates a histogram for numeric columns
* `plot.line`: Creates a line plot. Plots index on x column, and numeric column values for y
* `plot.pie`: Creates a pie plot

Additional plot types and plot functionality is available in the Seaborn package, which also supports pandas natively.
